```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('ValueIteration.csv', nrows=1000, delimiter=',')

for col in df.columns:
    plt.plot(df[col], label=col)

plt.legend(bbox_to_anchor =(0.5,-0.50), loc='lower center',ncol=6)
plt.show()
```
